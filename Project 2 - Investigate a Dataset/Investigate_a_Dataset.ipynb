## 1. Data Gathering 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

df= pd.read_csv('Database_TMDb_movie_data/tmdb-movies.csv')
```

## 2. Data Cleaning


```python
#df.sample(3)
#df.info()
#df.isnull().sum(), 
#df['genres'].value_counts()
#df['runtime'].describe()
#df['vote_count'].describe()
df.drop(['homepage','release_date','overview','tagline', 'keywords'], axis=1, inplace=True)
df.sample(4)
```

> 1. I believe the following columns won't be useful for the analysis:
    - homepage
    - release_date
    - overview
    - tagline
    - keywords (`genres` column summarizes the keywords column)
 - ######  As they are irrelevant to answer the questions posed in the `Introduction/Analysis Questions` section

```python
df_clean_c=df.copy()                                        #Collecting info about genres, cast, direcotors and cast columns
df_clean_c.dropna(inplace=True, subset=['cast'])            #While deleting NaN values to avoid errors
df_clean_c['cast']= df_clean_c['cast'].apply(lambda x: x.split('|'))
cast_dict={}
for cast in df_clean_c['cast']:
    for c in cast:
        if c not in cast_dict:
            cast_dict[c]=1
        if c in cast_dict:
            cast_dict[c]+=1            
df_clean_cd=df_clean_c.copy()     
df_clean_cd.dropna(inplace=True,subset=['director'])
df_clean_cd['director']=df_clean_cd['director'].apply(lambda x: x.split('|'))
directors_dict={}
for directors in df_clean_cd['director']:
    for d in directors:
        if d not in directors_dict:
            directors_dict[d] =1
        if d in directors_dict:
            directors_dict[d]+=1
df_clean_cdg=df_clean_cd.copy()                                            #Trying to lessen the ammount of removed rows of             
df_clean_cdg.dropna(inplace=True,subset=['genres'])                        #NaN to ensure high quality details 
df_clean_cdg['genres']=df_clean_cdg['genres'].apply(lambda x: x.split('|'))
genres_dict={}
for genre in df_clean_cdg['genres']:
    for g in genre:
        if g not in genres_dict:
            genres_dict[g] =1
        if g in genres_dict:
            genres_dict[g]+=1
df_clean_cdgp=df_clean_cdg.copy()    
df_clean_cdgp.dropna(inplace=True, subset=['production_companies']) #I dropped NaN values here, due to the huge number of
df_clean_cdgp['production_companies']=df_clean_cdgp['production_companies'].apply(lambda x: x.split('|'))
companies_dict={}                                    #NaN values present in this column, so it would not 
for companies in df_clean_cdgp['production_companies']:       #affect the info got from genres/directors/cast .
    for c in companies:
        if c not in companies_dict:
            companies_dict[c]= 1
        if c in companies_dict:
            companies_dict[c]+=1   
            
companies_dict2={}    
for companies in df_clean_cdgp['production_companies']:       #affect the info got from genres/directors/cast .
    for c in companies:
        if c not in companies_dict2:
            companies_dict2[c]= df_clean_cdgp[df_clean_cdgp['production_companies']==c]['vote_average'].sum()
        if c in companies_dict2:
            companies_dict2[c]+=df_clean_cdgp[df_clean_cdgp['production_companies']==c]['vote_average'].sum()  

                                                            #I didn't use a def function due to the number of modified variables 
                                                            #I needed like df_clean_c/cd/cdg/cdgp. I made them specifically to
                                                            #Avoid deleting useful info in the rows about to be deleted.
genres_tuple=tuple(genres_dict)
directors_tuple=tuple(directors_dict)
companies_tuple=tuple(companies_dict)
cast_tuple=tuple(cast_dict)

def top3(dict_name):
    for top in sorted(dict_name.values())[-3:]:
        x=print(list(dict_name.keys())[list(dict_name.values()).index(top)])
    return x
```

- I left null rows of `imdb_id` as they are, due to no correlation between other columns. Same goes for many other rows of different columns, except for `cast` so to append it to the `cast_list`, same goes for other newly created lists.

- ### Q.9 (The relation between `release_year` and ( [`budget_adj`, `revenue_adj`],`popularity`, `vote_average`)

```python
df['change_adj']= df['revenue_adj']-df['budget_adj']
df.groupby(['release_year'])['budget_adj', 'revenue_adj', 'change_adj'].mean().plot(legend=True)
plt.xlabel('Movie Release Year')
plt.ylabel('Movie Budget / Movie Revenue')
plt.title('Time VS Revenue & Budget')
plt.rcParams['figure.figsize'] = [18, 10]; 
```

- ##### The Time after 1970 till 1977 was the golden age for making revenue in the movie industry
- #### The Dataset correlation between revenue & budget at 2015 show that revenue is about double the budget

```python
df_3=df[df['popularity']> 3]                                 #I removed extra pop. from nill to three to ensure accuracy
sns.regplot(df_3['release_year'], df_3['popularity']);
```

- #### Popularity increases with time
- #### Popularity is more focused at the time period of 2000 to 2015

```python
df_4=df[df['vote_count']> 200]
df_4.groupby(['release_year'])['vote_average'].mean().plot(legend=True, kind='line')
plt.xlabel('Movie Release Year')
plt.ylabel('Vote Average')
plt.title('Vote-Average VS Time');
```

- #### The vote_average seems to be declining with time
- #### It seems like the time period between 2000 to 2015 has been its lowest, so contribution into making a better vote-average will have a great effect in the movie industry

- ### Q.12 (What affects `Revenue_adj`?)

```python
dff= df[df['revenue_adj']>0]                               #Excluded revenue with nill values for better visualisation
x= dff.groupby(['id']).median().budget_adj
y= dff.groupby(['id']).median().change_adj
sns.regplot(x, y);
```

- #### There is no defined correlation, but it seems that budgets from 1.7 to 2 are the best choices if the movie fails

```python
x= dff.groupby(['id']).median().vote_average
y= dff.groupby(['id']).median().revenue_adj
sns.regplot(x, y);
```

- #### Again, there is no strong defined correlation, but it seems that the more the average-vote the more likely the revenue is high

```python
x= dff.groupby(['id']).median().runtime
y= dff.groupby(['id']).median().change_adj
sns.regplot(x, y);
```

- #### Looks like the regression line show that the more runtime the more the revenue, but to be safe a 100 to 200 would be a good choice due to its affect on average-vote

```python
x= dff.groupby(['id']).median().popularity
y= dff.groupby(['id']).median().revenue_adj
sns.regplot(x, y);
```

- #### The regression line show that a popularity of 4 to 6 correlate with more revenue than other ranges

- Q.12) Revenue is affected by those descendingly:
    1. Popularity
    2. Runtime
    3. Budget
    4. Vote-Average
